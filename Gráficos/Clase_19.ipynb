```python
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
valoresx = np.random.randint(0,100,50)
valoresy = np.random.randint(0,100,50)
```

```python
datos = [go.Scatter(x=valoresx, y=valoresy, mode='markers')]
```

```python
pyo.plot(datos, filename='scatter.html')
```

```python
capa = go.Layout(title='Scatter plot', xaxis=dict(title='Valores x'), yaxis=dict(title='valores y'))
```

```python
figura = go.Figure(data = datos, layout=capa)
```

```python
pyo.plot(figura, filename='scatter.html')
```

```python
datos = [go.Scatter(x=valoresx, y=valoresy, mode='markers', 
                    marker=dict(size=10, color='blue', symbol='square', line=dict(width=2)))]
```

```python
capa = go.Layout(title='Scatter plot', xaxis=dict(title='Valores x'), yaxis=dict(title='valores y'))
figura = go.Figure(data = datos, layout=capa)
pyo.plot(figura, filename='scatter.html')
```
